### Part (a) Convolutional Network - 5 pt

Build a convolutional neural network model that takes the (224x224 RGB) image as input, and predicts the gesture
letter. Your model should be a subclass of nn.Module. Explain your choice of neural network architecture: how
many layers did you choose? What types of layers did you use? Were they fully-connected or convolutional?
What about other decisions like pooling layers, activation functions, number of channels / hidden units?

```python
class ASLCNN(nn.Module):
    def __init__(self):
        super(ASLCNN, self).__init__()
        self.conv1 = nn.Conv2d(3, 5, 5) #in_channels, out_chanels, kernel_size
        self.pool = nn.MaxPool2d(2, 2) #kernel_size, stride 
        self.conv2 = nn.Conv2d(5, 10, 5) #in_channels, out_chanels, kernel_size
        self.fc1 = nn.Linear(53*53*10, 32)
        self.fc2 = nn.Linear(32, 9)

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = x.view(-1, 53*53*10)
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return x

'''
I chose two convolutional networks and two fully connected layers. the reason I
chose two convolutional networks is because when looking at our specific problem, 
the AI only needs to recognize 2 key features, vertical and horizontal edge.
With this, I felt like there was enough information for the AI to figure out which
letter was being shown. I used two fully connected layers, the first to take in raw
inputs and use as a buffer to make the AI have more space to work with. I chose 
32 as the next layer as 224x224 is easily divisble by 32. 

I started with 5 output channels since it allows information to split into 5 easily. 
Then i grew the output to 10 since I wanted the convolutional network to have more
than enough information. I used maxpool since I want the network to detect edges which
requires the max of the kernel. I made the kernel 5 since no edge on a hand would 
be greater than 5 pixels wide. I chose relu since we can ignore negative values
as that would provide less important information to the ai. 
'''
```

```python
class ASLCNN2(nn.Module):
    def __init__(self):
        super(ASLCNN2, self).__init__()
        self.conv1 = nn.Conv2d(3, 10, 5) #in_channels, out_chanels, kernel_size
        self.pool = nn.MaxPool2d(2, 2) #kernel_size, stride 
        self.conv2 = nn.Conv2d(10, 30, 9) #in_channels, out_chanels, kernel_size
        self.fc1 = nn.Linear(102*102*30, 64)
        self.fc2 = nn.Linear(64, 9)

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = x.view(-1, 102*102*30)
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return x
```

```python
def train(model, train_loader, val_loader, batch_size=27, num_epochs=1, learn_rate = 0.001):
  torch.manual_seed(1000)
  criterion = nn.CrossEntropyLoss()
  optimizer = optim.Adam(model.parameters(), lr=learn_rate)
  losses, train_acc, val_acc, iters = [], [], [], []
  start_time = time.time()
  # training
  print ("Training Started...")
  n = 0 # the number of iterations
  for epoch in range(num_epochs):
    print('epoch #: ' + str(epoch))
    for imgs, labels in iter(train_loader):      
      #############################################
      #To Enable GPU Usage
      if torch.cuda.is_available():
        imgs = imgs.cuda()
        labels = labels.cuda()
      #############################################
      out = model(imgs)             # forward pass
      loss = criterion(out, labels) # compute the total loss
      loss.backward()               # backward pass (compute parameter updates)
      optimizer.step()              # make the updates for each parameter
      optimizer.zero_grad()         # a clean up step for PyTorch
      
    iters.append(n)
    losses.append(float(loss)/batch_size)             # compute *average* loss
    train_acc.append(get_accuracy(model, train_loader)) # compute training accuracy 
    val_acc.append(get_accuracy(model, valid_loader))  # compute validation accuracy
    print('train acc: ' + str(train_acc[-1]) + ' | train loss: ' + str(float(loss)) + ' | valid acc: ' + str(val_acc[-1]))
    n+=1

  print('Finished Training')
  end_time = time.time()
  elapsed_time = end_time - start_time
  print("Total time elapsed: {:.2f} seconds".format(elapsed_time))
  # plotting
  plt.title("Training Curve")
  plt.plot(iters, losses, label="Train")
  plt.xlabel("Iterations")
  plt.ylabel("Loss")
  plt.show()

  plt.title("Training Curve")
  plt.plot(iters, train_acc, label="Train")
  plt.plot(iters, val_acc, label="Validation")
  plt.xlabel("Iterations")
  plt.ylabel("Training Accuracy")
  plt.legend(loc='best')
  plt.show()

  print("Final Training Accuracy: {}".format(train_acc[-1]))
  print("Final Validation Accuracy: {}".format(val_acc[-1]))
  return train_acc, val_acc


def get_accuracy(model, data_loader):
    correct = 0
    total = 0
    for imgs, labels in data_loader:
        
        if torch.cuda.is_available():
          imgs = imgs.cuda()
          labels = labels.cuda()

        output = model(imgs)
        #select index with maximum prediction score
        pred = output.max(1, keepdim=True)[1]
        correct += pred.eq(labels.view_as(pred)).sum().item()
        total += imgs.shape[0]
    return correct / total


'''
I chose the Adam optimizer because it is very good at minimizing the error. 
It incorporates momentum therefore it is able to avoid local minimums and
saddle points. Also, the learning rate and momentim is adjusted for the
training set. So, if the error approaches a global minimum (after training for a
long time), it will adjust its learning rate and momentum so it will 
not skip over it. 

I chose cross enropy loss since you are able to acheive a perfect loss of 0. 
Also, since it finds the difference between the label and the predicted value
and adjusts the weights accordingly, it makes it a very good loss function.
'''
```

```python
#tensor = torch.from_numpy(tensor.detach().numpy())

def train_alex(model, batch_size, num_epochs, lr, train_features_arr, valid_features_arr, train_labels, valid_labels):
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=lr)

    iters, losses, train_acc, val_acc = [], [], [], []

    # training
    n = 0 # the number of iterations
    print(num_epochs)
    for epoch in range(num_epochs):
      print('epoch #" ' + str(epoch))
      for i in range(iter(train_features_arr)):
        if torch.cuda.is_available():
          train_features_arr[i] = train_features_arr[i].cuda()
          train_labels[i] = train_labels[i].cuda()
        out = model(train_features_arr[i])             # forward pass
        loss = criterion(out, train_labels[i]) # compute the total loss
        loss.backward()               # backward pass (compute parameter updates)
        optimizer.step()              # make the updates for each parameter
        optimizer.zero_grad()         # a clean up step for PyTorch

      # save the current training information
      iters.append(n)
      losses.append(float(loss)/batch_size)             # compute *average* loss
      train_acc.append(get_accuracy_alex(model, train_features_arr)) # compute training accuracy 
      val_acc.append(get_accuracy_alex(model, valid_features_arr))  # compute validation accuracy
      print('train acc: ' + str(train_acc[n]) + ' | train loss: ' + str(float(loss)) + ' | valid acc: ' + str(val_acc[n]))
      n += 1

    # plotting
    plt.title("Training Curve")
    plt.plot(iters, losses, label="Train")
    plt.xlabel("Iterations")
    plt.ylabel("Loss")
    plt.show()

    plt.title("Training Curve")
    plt.plot(iters, train_acc, label="Train")
    plt.plot(iters, val_acc, label="Validation")
    plt.xlabel("Iterations")
    plt.ylabel("Training Accuracy")
    plt.legend(loc='best')
    plt.show()

    print("Final Training Accuracy: {}".format(train_acc[-1]))
    print("Final Validation Accuracy: {}".format(val_acc[-1]))
    
def get_accuracy_alex(model, features_arr, labels):
  correct = 0
  total = 0
  for i in range(len(features_arr)):
      
      #############################################
      #To Enable GPU Usage
      if torch.cuda.is_available():
        features_arr[i] = features_arr[i].cuda()
        labes[i] = labes[i].cuda()
      #############################################
      output = model(features_arr[i])
      
      #select index with maximum prediction score
      pred = output.max(1, keepdim=True)[1]
      correct += pred.eq(labels[i].view_as(pred)).sum().item()
      total += imgs.shape[0]
  return correct / total
```
