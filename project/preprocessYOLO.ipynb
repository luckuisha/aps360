```
name = 'validation'

data = pd.read_csv(f'clean-{name}-id.csv')
XCenter = (data['XMax'] + data['XMin']) / 2
YCenter = (data['YMax'] + data['YMin']) / 2
width = (data['XMax'] - data['XMin'])
height = (data['YMax'] - data['YMin']) 

data['XMin'] = XCenter
data['XMax'] = YCenter
data['YMin'] = width
data['YMax'] = height
columns = ["filePath", "Class_Description", "XCenter", "YCenter", "Width", "Height"]
data.set_axis(columns, axis='columns', inplace=True)
```
